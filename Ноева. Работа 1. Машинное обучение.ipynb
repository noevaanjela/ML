**Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:**

```python
# Корреляционная матрица
numeric_df = df.select_dtypes(include=['int64', 'float64'])
numeric_df = [col for col in numeric_df if col not in ['Weekdaysort','Monthsort']]
correlation_matrix = df[numeric_df].corr()
    # Визуализация тепловой карты
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```
